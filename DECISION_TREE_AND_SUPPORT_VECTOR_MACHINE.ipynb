```python
import numpy as np

# Define a function for standardization
def standardize_data(data):
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    standardized_data = (data - mean) / std
    return standardized_data

# Standardize the training and test sets
X_train = standardize_data(X_train)
X_test = standardize_data(X_test)
```

# ***2. Decision tree model with entropy implementation***

## **_2.1 Implementation of the Model_**

```python
import numpy as np
import pandas as pd

class Node:
    def __init__(self, feature_index=None, threshold=None, value=None, left=None, right=None):
        self.feature_index = feature_index
        self.threshold = threshold
        self.value = value
        self.left = left
        self.right = right

class DecisionTreeClassifier:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.tree = None

    def fit(self, X, y):
        self.tree = self._build_tree(X, y, depth=0)

    def _build_tree(self, X, y, depth):
        num_samples, num_features = X.shape
        unique_classes, counts = np.unique(y, return_counts=True)

        if len(unique_classes) == 1 or (self.max_depth is not None and depth == self.max_depth):
            return Node(value=unique_classes[np.argmax(counts)])

        best_feature_index, best_threshold = self._find_best_split(X, y)

        left_mask = X[:, best_feature_index] < best_threshold
        right_mask = ~left_mask
        left_subtree = self._build_tree(X[left_mask], y[left_mask], depth + 1)
        right_subtree = self._build_tree(X[right_mask], y[right_mask], depth + 1)

        return Node(feature_index=best_feature_index, threshold=best_threshold, left=left_subtree, right=right_subtree)

    def _find_best_split(self, X, y):
        num_samples, num_features = X.shape
        classes = np.unique(y)
        best_gini = float('inf')
        best_feature_index = None
        best_threshold = None

        for feature_index in range(num_features):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                left_mask = X[:, feature_index] < threshold
                right_mask = ~left_mask
                gini = self._calculate_gini(y[left_mask], y[right_mask])

                if gini < best_gini:
                    best_gini = gini
                    best_feature_index = feature_index
                    best_threshold = threshold

        return best_feature_index, best_threshold

    def _calculate_gini(self, left_labels, right_labels):
        left_size = len(left_labels)
        right_size = len(right_labels)
        total_size = left_size + right_size

        if total_size == 0:
            return 0

        p_left = left_size / total_size
        p_right = right_size / total_size

        gini_left = 1.0 - sum(((left_labels == c).sum() / left_size) ** 2 for c in np.unique(left_labels))
        gini_right = 1.0 - sum(((right_labels == c).sum() / right_size) ** 2 for c in np.unique(right_labels))

        gini = p_left * gini_left + p_right * gini_right

        return gini

    def _predict_sample(self, x, node):
        if node.value is not None:
            return node.value

        if x[node.feature_index] < node.threshold:
            return self._predict_sample(x, node.left)
        else:
            return self._predict_sample(x, node.right)

    def predict(self, X):
        return np.array([self._predict_sample(x, self.tree) for x in X])

# Function for binning the target variable
def bin_target(y, num_bins):
    return pd.cut(y, bins=num_bins, labels=np.arange(num_bins)).astype(int)


for num_bins in [5]:
    y_train_bins = bin_target(y_train, num_bins)
    y_test_bins = bin_target(y_test, num_bins)

    # Instantiate and fit the Decision Tree classifier model
    decision_tree_classifier = DecisionTreeClassifier(max_depth=None)
    decision_tree_classifier.fit(X_train, y_train_bins)

    # Make predictions on the test set
    y_pred_bins = decision_tree_classifier.predict(X_test)

    # Evaluate the model using classification metrics
    accuracy_value = np.mean(y_test_bins == y_pred_bins)
    conf_matrix = np.zeros((num_bins, num_bins), dtype=int)

    for i in range(len(y_test_bins)):
        conf_matrix[y_test_bins[i], y_pred_bins[i]] += 1

    # Print the results for each binning strategy
    print(f"\nResults for {num_bins} bins:")
    print("Classification Accuracy:", accuracy_value)


    accuracy = np.mean(y_pred_bins == y_test_bins)
    print("Accuracy: {:.2f}%".format(accuracy * 100))
```

#***3. Adaboost***

## **_3.1 Implementation of the Model_**

```python
import numpy as np
import pandas as pd

class DecisionStump:
    def __init__(self):
        self.feature_index = None
        self.threshold = None
        self.alpha = None
        self.predictions = None

def predict_stump(X, feature_index, threshold):
    return np.where(X[:, feature_index] < threshold, -1, 1)

class AdaBoostClassifier:
    def __init__(self, n_estimators=50):
        self.n_estimators = n_estimators
        self.stumps = []
        self.alphas = []

    def fit(self, X, y):
        m, n = X.shape
        w = np.ones(m) / m

        for _ in range(self.n_estimators):
            stump = DecisionStump()

            min_error = float('inf')
            for feature_index in range(n):
                thresholds = np.unique(X[:, feature_index])
                for threshold in thresholds:
                    predictions = np.where(X[:, feature_index] < threshold, -1, 1)
                    error = np.sum(w * (predictions != y))

                    if error < min_error:
                        min_error = error
                        stump.feature_index = feature_index
                        stump.threshold = threshold
                        stump.predictions = predictions.copy()

            # Check if all predictions are the same
            if np.all(stump.predictions == stump.predictions[0]):
                stump.alpha = 0
            else:
                stump.alpha = 0.5 * np.log((1 - min_error) / max(min_error, 1e-10))

            # Update weights
            w = w * np.exp(-stump.alpha * y * stump.predictions)
            w /= np.sum(w)

            self.stumps.append(stump)
            self.alphas.append(stump.alpha)

    def predict(self, X):
        m = X.shape[0]
        predictions = np.zeros(m)

        for stump, alpha in zip(self.stumps, self.alphas):
            stump_pred = np.where(X[:, stump.feature_index] < stump.threshold, -1, 1)
            predictions += alpha * stump_pred

        return np.sign(predictions)

# Binning the target variable
num_bins = 5  # Adjust the number of bins as needed
y_train_bins = pd.cut(y_train, bins=num_bins, labels=np.arange(num_bins))
y_test_bins = pd.cut(y_test, bins=num_bins, labels=np.arange(num_bins))

# Convert category labels to integers
y_train_bins = y_train_bins.astype(int)
y_test_bins = y_test_bins.astype(int)

# Instantiate and fit the AdaBoost classifier model
adaboost_classifier = AdaBoostClassifier(n_estimators=50)
adaboost_classifier.fit(X_train_pca, y_train_bins)

# Make predictions on the test set
y_pred_bins = adaboost_classifier.predict(X_test_pca)

#classification metrics
accuracy = np.mean(y_test_bins == y_pred_bins)


# Print the classification metrics
print("Classification Accuracy:", accuracy)

accuracy = np.mean(y_pred_bins == y_test_bins)
print("Accuracy: {:.2f}%".format(accuracy * 100))

```

## **_4.1 Implementation of the Model_**

```python
import numpy as np




# SVM parameters
C = 1.0  # regularization parameter
learning_rate = 0.01  # learning rate
epochs = 1000  # number of training epochs

# SVM training using stochastic gradient descent
weights = np.zeros(X_train_pca.shape[1])
bias = 0

for epoch in range(epochs):
    for i in range(len(X_train_pca)):
        if y_train[i] * (np.dot(X_train_pca[i], weights) + bias) < 1:
            weights = weights - learning_rate * (2 * C * weights - np.dot(X_train_pca[i], y_train_bins[i]))
            bias = bias + learning_rate * y_train_bins[i]

# Make predictions on the test set
y_pred = np.sign(np.dot(X_test_pca, weights) + bias)

# Evaluate accuracy
accuracy = np.mean(y_pred == y_test_bins)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
